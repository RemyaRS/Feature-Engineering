# Importing Libraries

```python
import numpy as np
import pandas as pd

import warnings
# Filter warnings
warnings.filterwarnings('ignore')
```

#Making a dataset

```python
df_list= pd.DataFrame({'Roll_Number':[11, 14, 21, 24],
                  'Name':['Surya Chandra','Hamza Sheik','Jenifer Jos','Gurpreet Kaur'],
                  'Age' :[34, 28, 24, 39],
                  'gmail' :['unni@gmail.com', 'hamz@yahoo.com', 'jeny@gmail.com', 'guddi@gmail.com'],
                  'Subjects':['Data Analysis','Data Science','Data Engineering','Data Mining']})
```

```python
print(df_list)
```

Save the existing dataset in a separate variable

```python
df1=df_list
```

Split the column - Name, and save it in a new variable

```python
df2 = df_list.Name.str.split(expand=True)  
print(df2)
```

Rename the columns of dataset df2

```python
df2.columns = ['First Name', 'Last Name']
df2
```

Now combine the two dataset df1 and df2

```python
df=pd.concat([df1, df2], axis=1)
```

```python
df
```

# Splitting a Single Column

```python
import pandas as pd
# create a dataframe
data = pd.DataFrame({
    'Address': ['387, Chennai,Tamil Nadu',
                '305, New Delhi,Delhi',
                '782, Kolkata,West Bengal',
                '981, Mumbai,Maharashtra']
})
# display the dataframe
data
```

```python
# split column into multiple columns by delimiter 
data=data['Address'].str.split(',', expand=True)
data
```

```python
data.columns = ['Street Number', 'City', 'State']
data
```
